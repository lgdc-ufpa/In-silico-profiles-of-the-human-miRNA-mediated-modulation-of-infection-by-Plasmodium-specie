```python
# from youtube video
sns.set(rc={'figure.figsize':(11.7, 8.27)})
# color_bar_legend = 'Number of times that a micro-RNA was predicted.'
color_bar_legend = 'Targets'


sns.set(font_scale=0.8)
df_temp = df.pivot('Micro-RNAs', 'Species', 'Targets').fillna(0)

di_colummns = {}
for columm in df_temp.columns.values:
    li_columm = list(columm)
    li_columm[columm.index(' ')] = '\n'    
    di_colummns[columm] = "".join(li_columm)

di_colummns
df_temp.rename(columns=di_colummns, inplace=True)

# sns.heatmap(df_temp)
# lut = dict(zip(df_temp['Species'].unique(), "rbg"))
# row_colors = df_temp.map(lut)
# sns.clustermap(df_temp,  row_cluster=False, yticklabels=False, row_colors=lut)

################################
clustermap = sns.clustermap(df_temp,  
                            row_cluster=False, 
                            yticklabels=False,
                            cbar_kws={'label': color_bar_legend},
                           dendrogram_ratio=(.1, .2))

#---------------------------------#
"""
cbar position
https://stackoverflow.com/questions/46246345/moving-title-above-the-colorbar-in-seaborn-heatmap
"""
clustermap2 = sns.clustermap(df_temp.T,
#                              figsize=(7, 5),
#                              cbar_pos=(.0, 1, 0, .2),
                             col_cluster=False, 
                             xticklabels=False,
                             cbar_kws={'label': color_bar_legend,
                                      "orientation": "horizontal"})
#################################
# clustermap
clustermap2

# save figure
path = os.path.join(os.getcwd(), 'output', color_bar_legend + '.png')
# plt.ylabel('Species')
# plt.xlabel('Micro-RNAs')
plt.savefig(f'{path}', dpi=500, bbox_inches='tight')
plt.savefig(f'{path}', dpi=500)
plt.show()
plt.clf()
plt.close()
```
